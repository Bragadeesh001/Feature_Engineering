## One hot encoding with Many categories - nominal

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X0', 'X1', 'X2' ,'X3', 'X4','X5', 'X6'])
df.head()
```

```python
df['X0'].value_counts()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
for i in df.columns:
    print(df[i].value_counts())
```

```python
top10=df.X1.value_counts().sort_values(ascending=False).index[:10]
top10=list(top10)
top10
```

```python
import numpy as np
for categories in top10:
    df[categories]=np.where(df["X1"]==categories,1,0)
```

```python
df
```

```python
#### lets do for all column at a time
```

```python
pd.set_option('display.max_columns',None)
df=pd.read_csv('mercedesbenz.csv',usecols=['X0', 'X1', 'X2' ,'X5'])
df.head()
```

```python
df.size
```

```python
for feature in df.columns:
    top10=[x for x in df[feature].value_counts().sort_values(ascending=False).head(10).index]
    for categories in top10:
        df[feature+"_"+categories]=np.where(df[feature]==categories,1,0)
```

```python
df
```

```python
df.drop(['X0','X1', 'X2', 'X5'],axis=1,inplace=True)
df
```

## Target Guided Ordinal Encoding
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
data=pd.read_csv('Titanic_train.csv',usecols=['Cabin','Survived'])
```

```python
data.head()
```

```python
data['Cabin'].fillna('Z')
```

```python
## here c85,c123 are seat numbers in c block
## replacing them with first letter

data['Cabin']=data['Cabin'].astype(str).str[0]
```

```python
data.head()
```

```python
data['Cabin'].unique()
```

```python
# in each column how many people got survived
# percentage of how many people got survived

data.groupby(['Cabin'])['Survived'].mean()
```

```python
data.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=data.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
data['Cabin_ordinal_labels']=data['Cabin'].map(ordinal_labels2)
```

```python
data.head()
```

## Mean Encoding


```python
data.groupby(['Cabin'])['Survived'].mean()
```

```python
data.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
## im mean encoding we replace 'A' with 0.46666666

col=data.groupby(['Cabin'])['Survived'].mean().to_dict()
col
```

```python
data['Cabin_meanEncoding']=data['Cabin'].map(col)
```

```python
data.head()
```

#### Advantages
1. It capture information within the label therefore rendering more predictive features
2. It create monotonic relationship between the target and variable

#### Disadvantages
1. It prones to overfitting

## Probability Ratio Encoding
1. Probability of Survived based on Cabin--- Categorical feature
2. Probability of Not Survived---- 1-pr(Survived)
3. pr(Survived)/pr(Died)
4. Dictionary to map cabin with probability
5. replace with the categorical feature

```python
data=pd.read_csv('Titanic_train.csv', usecols=['Cabin','Survived'])
data.head()
```

```python
data.Cabin.fillna('Missing',inplace=True)
```

```python
data.head()
```

```python
data.Cabin.unique()
```

```python
data['Cabin']=data['Cabin'].astype(str).str[0]
```

```python
data.head()
```

```python
data.Cabin.unique()
```

```python
prob_df=data.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df
```

```python
# formula is
# probability of Survived divided by probability of Died

prob_df['Ratio']=prob_df['Survived']/prob_df['Died']
prob_df
```

```python
probability_encoded=prob_df['Ratio'].to_dict()
```

```python
data['cabin_encoded']=data['Cabin'].map(probability_encoded)
```
